```python
def computeHCF(a,b):
    smaller = b if a>b else a
    hcf = 1
    for i in range(1, smaller+1):
        if (a % i ==0) and (b % i ==0):
            hcf = i
    return hcf

num1 = 12
num2 = 14
print("H.C.F of {0} and {1} is: {2}".format(num1,num2, computeHCF(num1,num2)))
```

```python
def fact(n):
    if n < 0:
        return 'negative digits does not have fac'
    elif n == 0 or n ==1:
        return 1
    else:
        return n*fact(n-1)

n = int(input('enter num get fact='))
fact(n)
```

```python
def fibo(n):
    fibo_seq = [0,1]
    for i in range(2, n):
        fibo_seq.append(fibo_seq[i-1] + fibo_seq[i-1])
    return fibo_seq

n = int(input('enter num to get fibo='))
fibo(n)
```

```python
def fibo(n):
    fibo_seq = [0,1]
    for i in range(2,n):
        fibo_seq.append(fibo_seq[i-1] + fibo_seq[i-2])
    return fibo_seq

n = int(input('enter num to get fibo='))
fibo(n)
```

```python
# Now VVI topic 
# we will learn how to perform different join operations in pd
# the same way which we do in sql the different joins 
# like inner joins and outer joins etc
# let's manualy create 2 different data frames
# THIS IS CALLED MERGE DataFrames
import pandas as pd

temperature_df = pd.DataFrame({
    "city": ["mumbai", "delhi", "bangalore","hydrabad"],
    "temperature": [32, 45, 30, 40] 
})

temperature_df
```

```python
# THIS IS CALLED MERGE DataFrames

import pandas as pd
# here we will remove hydrabad to perform outer join
humidity_df = pd.DataFrame({
    "city": ["mumbai", "delhi", "bangalore"],
    "humidity": [68, 65, 75] 
})

humidity_df
```

```python
# now merge 2 dataframes with out index and perform inner join
df = pd.merge(temperature_df, humidity_df, on='city')
df
# it will perform inner join based on city
```

```python
# outer join
df = pd.merge(temperature_df, humidity_df, on = 'city', how='outer')
df
# here where we do not have value it will put not a num NAN
```
